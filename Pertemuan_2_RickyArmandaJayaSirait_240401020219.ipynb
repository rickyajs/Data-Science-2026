Jawaban Pertanyaan Q1

```python
# Import library yang dibutuhkan
import pandas as pd
import numpy as np

# Memuat dataset Titanic dari URL GitHub resmi di modul
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df =   pd.read_csv(url)

# Menggunakan atribut .shape untuk mendapatkan dimensi data (baris, kolom)
baris, kolom = df.shape
print(f"Jumlah Baris: {baris}")
print(f"Jumlah Kolom: {kolom}")
```

Jawaban Pertanyaan Q2

```python
# Menghitung jumlah missing values per kolom
missing_count = df.isnull().sum()

# Menghitung proporsi (%) missing values per kolom
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Menggabungkan hasil ke dalam satu DataFrame agar rapi
missing_data = pd.DataFrame({
    'Jumlah Missing': missing_count,
    'Proporsi (%)': missing_percentage
})

# Filter hanya kolom yang memiliki missing values > 0
missing_data_filtered = missing_data[missing_data['Jumlah Missing'] > 0]
print(missing_data_filtered.round(2))
```

Jawaban Pertanyaan Q3

```python
# Menghitung jumlah total (counts) untuk masing-masing kategori
jumlah_korban = df['Survived'].value_counts()

# Menghitung persentase (proporsi) untuk masing-masing kategori
persen_korban = df['Survived'].value_counts(normalize=True) * 100

# Menggabungkan keduanya ke dalam sebuah DataFrame baru agar rapi
tabel_q3 = pd.DataFrame({
    'Jumlah Orang': jumlah_korban,
    'Persentase (%)': persen_korban
})

# Mengubah nama indeks agar lebih mudah dibaca (0 -> Tidak Selamat, 1 -> Selamat)
tabel_q3.index = ['Tidak Selamat', 'Selamat',]

# Menampilkan tabel hasil
print(tabel_q3.round(2))
```

Jawaban Pertanyaan Q4

```python
# Melakukan filtering dengan kondisi ganda (Boolean Indexing)
# Kondisi: Pclass == 1 DAN Sex == 'female'
wanita_kelas1 = df[(df['Pclass'] == 1) & (df['Sex'] == 'female')]

# Menghitung jumlah baris hasil filter
jumlah_wanita_kelas1 = len(wanita_kelas1)

# Menghitung rata-rata usia dengan mengabaikan missing values secara otomatis
rata_usia_wanita_kelas1 = wanita_kelas1['Age'].mean()

print(f"Jumlah penumpang wanita dari Kelas 1: {jumlah_wanita_kelas1} orang")
print(f"Rata-rata usia mereka: {rata_usia_wanita_kelas1:.2f} tahun")
```

Jawaban Pertanyaan Q5

```python
# Mengelompokkan berdasarkan Pclass dan menghitung rata-rata dari kolom Survived
tingkat_keselamatan_per_kelas = df.groupby('Pclass')['Survived'].mean() * 100
print(tingkat_keselamatan_per_kelas.round(2))

# Mencari kelas dengan tingkat keselamatan tertinggi
kelas_tertinggi = tingkat_keselamatan_per_kelas.idxmax()
nilai_tertinggi = tingkat_keselamatan_per_kelas.max()

print(f"\nKelas penumpang dengan tingkat keselamatan tertinggi adalah Kelas {kelas_tertinggi} dengan {nilai_tertinggi:.2f}%")
```

Jawaban Pertanyaan Q6

```python
# Mengelompokkan berdasarkan Jenis Kelamin dan Kelas untuk melihat rata-rata keselamatan
analisis_gender = df.groupby(['Sex'])['Survived'].mean() * 100
print("Tingkat Keselamatan Berdasarkan Jenis Kelamin:")
print(analisis_gender.round(2))

print("\n" + "="*50 + "\n")

# Analisis kombinasi: Jenis Kelamin dan Kelas Penumpang
analisis_kombinasi = df.groupby(['Pclass', 'Sex'])['Survived'].mean() * 100
print("Tingkat Keselamatan Berdasarkan Kelas dan Jenis Kelamin:")
print(analisis_kombinasi.round(2).unstack())
```
