```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
print("=== MEMUAT DATASET ===")
df = sns.load_dataset('penguins')

# Menghapus missing values secara singkat agar visualisasi tidak terganggu nilai kosong
df = df.dropna()

print("Dataset 'penguins' berhasil dimuat.")
print(f"Dimensi data (baris, kolom): {df.shape}")
print(df.head())
print("-" * 60)
```

```python
print("=== MEMBUAT VISUALISASI DISTRIBUSI ===")

# Membuat Figure tunggal untuk Distribusi Panjang Paruh (Bill Length) per Spesies
plt.figure(figsize=(10, 6))
sns.histplot(
    data=df,
    x='bill_length_mm',
    hue='species',
    multiple='stack',
    kde=True,
    palette='Set2'
)
plt.title('Distribusi Panjang Paruh (Bill Length) Berdasarkan Spesies Penguin')
plt.xlabel('Panjang Paruh (mm)')
plt.ylabel('Hitungan (Count)')
plt.tight_layout()
plt.show()

print("-" * 60)
```

```python
print("=== SUBPLOTS HUBUNGAN VARIABEL & KUSTOMISASI ===")

# Inisialisasi kanvas subplots: 1 baris, 2 kolom sesuai arahan template gambar
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Scatter Plot (Hubungan Panjang Flapper vs Massa Tubuh)
sns.scatterplot(
    data=df,
    x='flipper_length_mm',
    y='body_mass_g',
    hue='species',
    style='island',
    palette='Set2',
    s=70,
    ax=axes[0]
)
axes[0].set_title('Hubungan Panjang Flapper vs Massa Tubuh', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Panjang Flapper (mm)', fontsize=10)
axes[0].set_ylabel('Massa Tubuh (g)', fontsize=10)

# Box Plot (Distribusi Massa Tubuh per Jenis Kelamin & Spesies)
sns.boxplot(
    data=df,
    x='species',
    y='body_mass_g',
    hue='sex',
    palette='Pastel1',
    ax=axes[1]
)
axes[1].set_title('Distribusi Massa Tubuh Berdasarkan Spesies & Jenis Kelamin', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Spesies', fontsize=10)
axes[1].set_ylabel('Massa Tubuh (g)', fontsize=10)

# Mengoptimalkan tata letak ruang antar grafik
plt.tight_layout()
plt.show()

print("-" * 60)

print("=== STRUKTUR MATRIKS KORELASI (FITUR NUMERIK) ===")

plt.figure(figsize=(8, 5))
numeric_cols = df.select_dtypes(include=[np.number]).columns
corr_matrix = df[numeric_cols].corr()

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    linewidths=0.5
)
plt.title('Heatmap Korelasi Fitur Numerik Dataset Penguin')
plt.tight_layout()
plt.show()
```
